## Berechnungen zum Bose-Gas

Durch numerische Integration wird das Integral für die Teilchendichte für festgelegte Temperaturen T berechnet. Dabei soll das Integral 1 ergeben, womit sich eine Bedingung für z(T) ergibt. (Dabei liegt z zwischen 0 und 1, weil chemische Potential <1 ist.) Somit kann man die Fugazität z(T) variieren und durch Intervallschachtelung immer weiter annähern. Darauf folgend wird das chemische potential µ(T) berechnet. Anschließend wird die Innere Energie E(T) durch Nummerische Integration für die vorherigen Temperaturen T und berechneten µ(T) ausgerechnet. Zum Schluss wurde noch eine numerische Ableitung der inneren Energie gebildet, um die Wärmekapazität zu erhalten.


Teilchendichte:
\begin{equation}
N(T) = \int_0^\infty \frac{\sqrt{E}}{e^{(E-µ)/T}-1} = \int_0^\infty \frac{\sqrt{E}ze^{-E/T}}{1-ze^{-E/T}} = 1
\end{equation}

Fugazität
\begin{equation}
z=e^{µ/T}
\end{equation}

chemisches Potential:
\begin{equation}
µ= \ln(z) \cdot T
\end{equation}

Innere Energie:
\begin{equation}
E(T) = \int_0^\infty \frac{\sqrt{E^3}}{e^{(E-µ)/T}-1} = \int_0^\infty \frac{\sqrt{E^3}ze^{-E/T}}{1-ze^{-E/T}}
\end{equation}

Wärmekapazität:
\begin{equation}
c_v(T) = \frac{\partial E}{\partial T} \Bigg|_V
\end{equation}


Unterhalb einer kritischen Temperatur $\mathrm{T_c} \approx 0.57 \pm 0.02$ ist das Teilchenzahlintegral kleiner 1, weil mehrere Teilchen den selben Zustand (den Grundzustand bei tiefen Temperaturen) besetzen.

\begin{equation}
N(T) = \int_0^\infty \frac{\sqrt{E}ze^{-E/T}}{1-ze^{-E/T}} < 1
\end{equation}

Somit lässt sicher der Anteil der Teilchen im Grundzustand so berechnen:

\begin{equation}
N_g(T) = 1 - N(T)
\end{equation}

```python
%reset -f
import numpy as np
import matplotlib.pyplot as plt

#functions:
fu = lambda E,T,z : np.sqrt(E)*z*np.exp(-E/T)/(1-z*np.exp(-E/T))    #particle count
fe = lambda E,T,z : np.sqrt(E**3)*z*np.exp(-E/T)/(1-z*np.exp(-E/T)) #internal energy


def integrate(f,T,z):
    
    ##########setup##########
    #get E array:
    N=1000000 #steps
    #set start & end for array:
    threshold=1e-8 #threshold
    a=1e-10
    #find b:
    bit=True;iii=5
    while bit==True:
        if f(iii,T,z)>threshold:
            iii=iii*1.2
        else:
            b=iii
            bit=False
    
    ##########integrate##########
    E=np.linspace(a,b,N) #grit
    dE=(b-a)/N #step
    value= dE*np.sum(f(E,T,z)) #sum: step * grit values
    
    
    #print('integrate: \n b=',b,'\n integral=',value) #///testing
    
    return value

def nested_interval_z(f,T):
    #calculate u from f(E,T,z) by nested interval approximation
    
    ##########setup##########
    result=1 #expected integration result
    #interval:
    n=0 #start
    m=1#.1 #end
    uncertainty=1e-9
    
    ##########nested interval approx.##########
    i=0
    while i<200:
        w=(m+n)/2
        #print('w=',w) #///testing
        resultw=integrate(f,T,w)
        '''
        if np.isnan(resultw)==True:
            #integral nan
            w=1            
            i=99999999
        '''
        if resultw < 1:
            n=w
        elif (resultw > 1) or (resultw==np.nan):
            m=w
        else:
            n=w*99/100000
            m=w*101/100000
            
        i=i+1
        
        if np.abs(resultw-1) < uncertainty :
            #ratio of particles in ground state
            i=99999999
    #print('result:\n T=',T,'\n z=',w,'\n integral=',resultw,'\n done \n') #///testing
    N= 1- resultw
        
    return w, N

###defined functions:
#integrate(f,T,z)
#nested_interval_z(f,T)


#fugacity & ratio of paticles in Groundstate
#obtain values from T0 to 2T_F with M steps
M=100
T0=1e-6
T=np.linspace(T0,3,M)
#T=np.logspace(np.log10(T0),np.log10(2*T_F),M)
zz=np.zeros(M)
Ng =np.zeros(M)
for j in range(0,M):
    zz[j],Ng[j]=nested_interval_z(fu,T[j])

#chemical potential
mue=np.log(zz)*T


#internal energy
Ee =np.zeros(M)
for ii in range(0,M):
    Ee[ii] = integrate(fe,T[ii],zz[ii])

#heat capacity
dT=T[1]-T[0] #delta T
cv = (np.delete(Ee, 0, 0)-np.delete(Ee, M-1,0))/dT #difference/dT


########################################plot


plt.figure(num=None, figsize=(10, 25), dpi=80, facecolor='w', edgecolor='k')



plt.subplot(511)
plt.scatter(T,zz, s=6, color='blue')

plt.title('Fugazität in Abhängigkeit von der Temperatur', fontsize=16)
plt.xlabel('T', fontsize=13)
plt.ylabel('z(T)', fontsize=13)
plt.grid(True)



plt.subplot(512)
plt.scatter(T,mue, s=6, color='blue')

plt.title('chemisches Potential in Abhängigkeit von der Temperatur', fontsize=16)
plt.xlabel('T', fontsize=13)
plt.ylabel('µ(T)', fontsize=13)
plt.grid(True)



plt.subplot(513)
plt.scatter(T, Ee, s=6, color='blue')
           
plt.title('innere Energie in Abhängigkeit von der Temperatur', fontsize=16)
plt.xlabel('T', fontsize=13)
plt.ylabel('E(T)', fontsize=13)
plt.grid(True)



plt.subplot(514)
plt.scatter(np.delete(T, M-1,0)+dT, cv, s=6, color='blue')
           
plt.title('Wärmekapazität in Abhängigkeit von der Temperatur', fontsize=16)
plt.xlabel('T', fontsize=13)
plt.ylabel('c_v(T)', fontsize=13)
plt.grid(True)


plt.subplot(515)
plt.scatter(T, Ng, s=6, color='blue')
           
plt.title('Anteil der Teilchen im Grundzustand', fontsize=16)
plt.xlabel('T', fontsize=13)
plt.ylabel('N_g(T)', fontsize=13)
plt.grid(True)


plt.tight_layout()
plt.show()
```
