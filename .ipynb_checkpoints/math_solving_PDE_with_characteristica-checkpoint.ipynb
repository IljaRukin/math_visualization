# solving PDE with method of characteristics

linear partial differential equation of second order can be rewritten as:

\begin{align}
\frac{\partial u}{\partial x} P(x,y,u) + \frac{\partial u}{\partial y} Q(x,y,u) &= R(x,y,u) \\
\Rightarrow \frac{\partial u}{\partial x} \frac{\partial x}{\partial \tau} + \frac{\partial u}{\partial y} \frac{\partial y}{\partial \tau} &=  \frac{\partial u}{\partial \tau} \\
\end{align}

by setting $P(x,y,u) = \frac{\partial x}{\partial \tau}$ and $Q(x,y,u) = \frac{\partial y}{\partial \tau}$, such that the left hand side becomes the partial derivative of $\frac{\partial u}{\partial \tau}$ wich is equal to $R(x,y,u)$.

---

example:
\begin{align}
4 \frac{\partial u}{\partial x} + 7 \frac{\partial u}{\partial y} &= 0 \\
P(x,y,u) = \frac{\partial x}{\partial \tau} &= 4 \\
Q(x,y,u) = \frac{\partial y}{\partial \tau} &= 7 \\
R(x,y,u) = \frac{\partial u}{\partial \tau} &= 0
\end{align}

from wich follows:
\begin{align}
x &= 4 \tau + c_1 \\
y &= 7 \tau + c_2 \\
u &= f(\xi)
\end{align}

this is effectively a coordinate transform $(x,y) \rightarrow (\tau,\xi)$, such that $u = f(\xi)$ is only dependent on one varible $\xi(x,y)$. In turn $u(\xi)$ is constant when varying variable $\tau$, therefore u is constant at $(x,y) = (4 \tau + c_1, 7 \tau + c_2)$ when varying only $\tau$. These lines $y = 7/4 x + \tilde{c}$ where u stays constant are called characteristica and constant $\tilde{c}$ shifts between them (for different $\tau$).

we add a boundry condition:
\begin{align}
u(x = 0,y) &= \sin(y)
\end{align}

since we now know the values of u at (x=0,y) and the function stays the same along the characteristica we get the complete solution as:
\begin{align}
u(x,y) &= \sin(y - 7/4 x)
\end{align}


```python
%reset -f
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib as mpl
from matplotlib import cm
```

```python
#excercise 13a

#####3d plane data
x = np.linspace(-4, 4, 90)
y = np.linspace(-4, 4, 90)
xx, yy = np.meshgrid(x, y)

boundry = np.sin(y)
solution = np.sin(yy-7/4*xx)
characteristica = lambda x,c:7/4*x +c

#####plot commands
fig = plt.figure(figsize=(12, 7))
ax = fig.add_subplot(111, projection='3d')
plt.title("PDE solution", fontsize=20)

surf = ax.plot_surface(xx,yy,solution, cmap=cm.coolwarm, linewidth=0, label="solution", zorder=1)
fig.colorbar(surf, shrink=0.5, aspect=5)
for i in range(-6,8,2):
    ax.plot(x,characteristica(x,i),np.linspace(0,0,90), c='orange', linewidth=1, zorder=0)
ax.plot(np.linspace(0,0,90),y,boundry, c='green', linewidth=4, zorder=10, label="boundry")

ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)
ax.set_zlim(-1,1)
ax.set_xticks([-4,-2,0,2,4])
ax.set_yticks([-4,-2,0,2,4])
ax.set_zticks([-1,0,1])
ax.set_xlabel('X axis', fontsize=12)
ax.set_ylabel('Y axis', fontsize=12)
ax.set_zlabel('Z axis', fontsize=12)
ax.view_init(elev=50, azim=60)
#ax.legend(bbox_to_anchor=(0.95,0.95))
plt.show()
```
