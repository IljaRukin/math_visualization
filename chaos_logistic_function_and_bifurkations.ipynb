### Logistic map
one iteration is calculates with following formula
\begin{equation}
x_{n+1} = r \cdot x_n (1 - x_n)
\end{equation}

<br>
Behavior for different values of r:
<br>
- $r \in [0,1]$ x diverges to zero<br>
- $r \in [1,2]$ x quickly diverges to $\frac{r - 1}{r}$<br>
- $r \in [2,3]$ x slowly diverges to $\frac{r - 1}{r}$<br>
- $r \in [3,1+\sqrt{6}]$ oscillation between two values<br>
- $r \in [1+\sqrt{6},3.54409]$ oscillation between four values<br>
... number of oscillation will double each time<br>

a bifurcation map to visualize this behaviour as well as some examples are plotted below


```python
%reset -f
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

global i,pi
i = np.complex(0,1)
pi = np.pi

##########calculate##########

N=100
num_x=51

fx = lambda x,r: r*(x - np.multiply(x,x))
x = np.zeros((N,num_x),dtype=float)

fig = plt.figure(num=None, figsize=(12,7), dpi=100, facecolor='w', edgecolor='k')
ax1=plt.subplot(111)

for r in iter(np.linspace(0,4,200)):
    x[0,:]=np.linspace(0,1,num_x)
    
    #discarded iterate
    for ii in range(0,N-1):
        x[0,:]=fx(x[0,:],r)
    
    #iterate
    for ii in range(0,N-1):
        x[ii+1,:]=fx(x[ii][:],r)

    for iii in range(0,num_x):
        plt.scatter(r*np.ones(N),x[:,iii], s=1, color='blue')

plt.title(r'bifurcation diagram of logistic map $x_{n+1} = r \cdot x_n (1 - x_n)$', fontsize=16)
plt.xlabel('r', fontsize=13)
plt.ylabel('x values', fontsize=13)
plt.grid(True)

#fig.gca().set_autoscale_on(False)
plt.tight_layout() #rescale subplots
plt.show()
```

```python
%reset -f
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

global i,pi
i = np.complex(0,1)
pi = np.pi

##########calculate##########

N=10
num_x=51
r=0.4

fx = lambda x,r: r*(x - np.multiply(x,x))
x = np.zeros((N,num_x),dtype=float)
x[0,:]=np.linspace(0,1,num_x)

#iterate
ii=0
for ii in range(0,N-1):
    x[ii+1,:]=fx(x[ii][:],r)

##########plot##########

fig = plt.figure(num=None, figsize=(12,7), dpi=80, facecolor='w', edgecolor='k')

ax1=plt.subplot(111)
for iii in range(0,num_x):
    plt.plot(x[:,iii], linewidth=2)
plt.title('logistic map plot r= '+str(r), fontsize=16)
plt.xlabel('iterations', fontsize=13)
plt.ylabel('x values', fontsize=13)
plt.grid(True)

#fig.gca().set_autoscale_on(False)
plt.tight_layout() #rescale subplots
plt.show()
```
