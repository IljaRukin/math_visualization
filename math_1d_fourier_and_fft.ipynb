## fourier transform

\begin{align}
\mathcal{F}\{rect(ax) \} = \frac{sin(\pi f / a)}{\pi f}
\end{align}

plot with exact calculation and fft

```python
#fourier transform
%reset -f
import numpy as np
import matplotlib.pyplot as plt

#step function
xmin,xmax = -4,4
N = 1024
Fs = (N-1)/(xmax-xmin) #samples per unit (second,meter,...)
dt = 1/Fs #time step
df = Fs/N #frequency step
x = np.linspace(xmin,xmax,N)
rect = lambda x: (np.abs(x)<0.5).astype(float)
a=1
y1 = rect(a*x)

#fourier transformed function
if N%2==1:
    #odd number of data samples
    f = np.linspace(-(Fs-df)/2,(Fs-df)/2,N)
else:
    #even number of data samples
    f = np.linspace(-(Fs)/2,(Fs-2*df)/2,N)
trafo = lambda x,a : np.sin(np.pi*x/a)/(np.pi*x)
y2 = trafo(f,a)
y3 = np.real(np.fft.fftshift(np.fft.fft(np.fft.fftshift(y1))))
y3 = y3 *np.sum(np.abs(y1))/np.sum(np.abs(y3)) * 1/(2*np.pi)

###plot commands
plt.figure(figsize=(12, 10), facecolor='w', edgecolor='k')

ax1 = plt.subplot(2,1,1)
plt.plot(x,y1,color="red",label="rect", linewidth=4)
plt.title("fourier trafo: rectangular function -> sinc function", fontsize=20)
plt.legend(fontsize=13, bbox_to_anchor=(0.9,0.8))
plt.xlabel("", fontsize=12)
plt.ylabel("", fontsize=12)
plt.xticks([-a/2,a/2], ('-a/2', 'a/2'))
plt.xlim(xmin,xmax)
plt.ylim(-1,2)

ax2 = plt.subplot(2,1,2)
plt.plot(f,y2,color="blue",label="sinc", linewidth=4)
plt.plot(f,y3,color="green",label="fft", linewidth=4)
plt.title("fourier trafo: rectangular function -> sinc function", fontsize=20)
plt.legend(fontsize=13, bbox_to_anchor=(0.9,0.8))
plt.xlabel("", fontsize=12)
plt.ylabel("", fontsize=12)
plt.xticks([-np.pi/a,np.pi/a], (r'$-\pi/a$', r'$\pi/a$'))
plt.xlim(-Fs/2,Fs/2)
plt.ylim(-1,2)

plt.show()
```

## fourier series

```python
#fourier series
%reset -f
import numpy as np
import matplotlib.pyplot as plt

#generate x values
xmin,xmax = -8,8
N = 1024
Fs = (N-1)/(xmax-xmin) #samples per unit (second,meter,...)
dt = 1/Fs #time step
df = Fs/N #frequency step
x = np.linspace(xmin,xmax,N)
sawtooth = lambda x,a: np.mod(x-a,2*a)-a
a=np.pi
y1 = sawtooth(x,a)

n1 = 6
n2 = 50

def fourierSum(n_max,x):
    sums = 0
    for n in range(1,n_max):
        sums += 2*(-1)**(n+1)/n*np.sin(n*x)
    return sums

y2_1 = fourierSum(n1,x)
y2_2 = fourierSum(n2,x)

if N%2==1:
    #odd number of data samples
    f = np.linspace(-(Fs-df)/2,(Fs-df)/2,N)
else:
    #even number of data samples
    f = np.linspace(-(Fs)/2,(Fs-2*df)/2,N)
series = lambda x,a : np.sin(np.pi*x/a)/(np.pi*x)
y3 = np.fft.fft(np.fft.fftshift(y1))
sort = np.sort(np.abs(y3))
mask_1 = np.abs(y3) >= sort[-n1*2]
mask_2 = np.abs(y3) >= sort[-n2*2]
y3_1 = np.fft.fftshift(np.fft.ifft( y3*mask_1 ))
y3_2 = np.fft.fftshift(np.fft.ifft( y3*mask_2 ))

###plot commands
plt.figure(figsize=(12, 12))

ax1 = plt.subplot(2,1,1)
plt.plot(x,y1,color="blue",label="Sawtooth wave fourier series approx", linewidth=4)
plt.plot(x,y2_1,color="red",label="Fourier series n="+str(n1), linewidth=3)
plt.plot(x,y2_2,color="green",label="Fourier series n="+str(n2), linewidth=3)
plt.title("Fourier Series approximation of a Sawtooth Function", fontsize=20)
plt.legend(fontsize=13, bbox_to_anchor=(0.8,0.25))
plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)

ax2 = plt.subplot(2,1,2)
plt.plot(x,y1,color="blue",label="Sawtooth wave fft approx", linewidth=4)
plt.plot(x,y3_1,color="red",label="Fourier series n="+str(n1), linewidth=3)
plt.plot(x,y3_2,color="green",label="Fourier series n="+str(n2), linewidth=3)
plt.title("FFT approximation of a Sawtooth Function", fontsize=20)
plt.legend(fontsize=13, bbox_to_anchor=(0.8,0.25))
plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)

plt.show()
```

## FFT

```python
#fft
%reset -f
import numpy as np
import matplotlib.pyplot as plt

#step function
N = 256
Fs = 64 #samples per unit (second,meter,...)
T = N/Fs #sample interval in units (second,meter,...)
dt = 1/Fs #time step
df = Fs/N #frequency step
x = np.arange(0,T,1/Fs)
rect = lambda x: np.sin(10*x)
a=1
y1 = rect(a*x)

#fourier transformed function
if N%2==1:
    #odd number of data samples
    f = np.linspace(-(Fs-df)/2,(Fs-df)/2,N)
else:
    #even number of data samples
    f = np.linspace(-(Fs)/2,(Fs-2*df)/2,N)
y2 = np.real(np.fft.fftshift(np.fft.fft(np.fft.fftshift(y1))))
y2 = y2 *np.sum(np.abs(y1))/np.sum(np.abs(y2)) * 1/(2*np.pi)

###plot commands
plt.figure(figsize=(12, 10), facecolor='w', edgecolor='k')

ax1 = plt.subplot(2,1,1)
plt.plot(x,y1,color="blue",label="func", linewidth=4)
plt.title("time domain", fontsize=20)
plt.legend(fontsize=13, bbox_to_anchor=(0.9,0.8))
plt.xlabel("", fontsize=12)
plt.ylabel("", fontsize=12)
plt.xlim(0,T)
plt.ylim(-1.2,1.2)

ax2 = plt.subplot(2,1,2)
plt.plot(f,y2,color="blue",label="fft", linewidth=4)
plt.title("frequency domain", fontsize=20)
plt.legend(fontsize=13, bbox_to_anchor=(0.9,0.8))
plt.xlabel("", fontsize=12)
plt.ylabel("", fontsize=12)
plt.xlim(-Fs/2,Fs/2)
plt.ylim(-1.5,4)

plt.show()
```

## discrete fourier transform

```python
#fft
%reset -f
import numpy as np
import matplotlib.pyplot as plt

i = np.complex(0,1)

#step function
N = 256
Fs = 64 #samples per unit (second,meter,...)
T = N/Fs #sample interval in units (second,meter,...)
dt = 1/Fs #time step
df = Fs/N #frequency step
x = np.arange(0,T,1/Fs)
rect = lambda x: np.sin(10*x)
a=1
y1 = rect(a*x)

#fourier transformed function
n=np.arange(N)
k=n.reshape((N,1))
myfft = lambda f: np.matmul(np.transpose(np.exp(-i*k*n*2*np.pi/N)),f)
if N%2==1:
    #odd number of data samples
    f = np.linspace(-(Fs-df)/2,(Fs-df)/2,N)
else:
    #even number of data samples
    f = np.linspace(-(Fs)/2,(Fs-2*df)/2,N)
y2 = np.real(np.fft.fftshift(myfft(np.fft.fftshift(y1))))
y2 = y2 *np.sum(np.abs(y1))/np.sum(np.abs(y2)) * 1/(2*np.pi)

###plot commands
plt.figure(figsize=(12, 10), facecolor='w', edgecolor='k')

ax1 = plt.subplot(2,1,1)
plt.plot(x,y1,color="blue",label="func", linewidth=4)
plt.title("time domain", fontsize=20)
plt.legend(fontsize=13, bbox_to_anchor=(0.9,0.8))
plt.xlabel("", fontsize=12)
plt.ylabel("", fontsize=12)
plt.xlim(0,T)
plt.ylim(-1.2,1.2)

ax2 = plt.subplot(2,1,2)
plt.plot(f,y2,color="blue",label="fft", linewidth=4)
plt.title("frequency domain", fontsize=20)
plt.legend(fontsize=13, bbox_to_anchor=(0.9,0.8))
plt.xlabel("", fontsize=12)
plt.ylabel("", fontsize=12)
plt.xlim(-Fs/2,Fs/2)
plt.ylim(-1.5,4)

plt.show()
```
