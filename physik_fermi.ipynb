## Berechnungen zum Fermi-Gas

Durch numerische Integration wird das Integral für die Teilchendichte für festgelegte Temperaturen T berechnet. Dabei soll das Integral 1 ergeben, womit sich eine Bedingung für µ(T) ergibt. Somit kann man das chemische Potential µ(T) variieren und durch Intervallschachtelung immer weiter annähern.
Anschließend wird die Innere Energie E(T) durch Nummerische Integration für die vorherigen Temperaturen T und berechneten µ(T) ausgerechnet.<br>

Teilchendichte:
\begin{equation}
N(T) = \int_0^\infty \frac{\sqrt{E}}{e^{(E-µ)/T}+1} = 1
\end{equation}

Innere Energie:
\begin{equation}
E(T) = \int_0^\infty \frac{\sqrt{E}^3}{e^{(E-µ)/T}+1}
\end{equation}

zum Vergleich: das chemische Potential entspricht bei $T=0 \approx 1e-6$ der Fermienergie, welche bei T=0 folgendermaßen gegeben ist:
\begin{equation}
µ(T=0)=E_F(T=0) = \biggl( \frac{3}{2} \biggr) ^{2/3} \approx 1.310371
\end{equation}


```python
%reset -f
import numpy as np
import matplotlib.pyplot as plt

#functions:
fu = lambda E,T,u : np.sqrt(E)/(np.exp((E-u)/T)+1)    #particle count
fe = lambda E,T,u : np.sqrt(E**3)/(np.exp((E-u)/T)+1) #internal energy


def integrate(f,T,u):
    #calcullate integral of function f(E,T,u) from 0 to infinity by summing up rectangles
    #integration start and end limit is chosen around the peak of the function to improve accuracy
    #parameter: T
    #variable: u
    
    ##########setup##########
    #get E array:
    N=1000000 #steps
    #set start & end for array:
    threshold=1e-8 #threshold
    a=0
    #find b:
    bit=True;iii=1
    while bit==True:
        if f(iii,T,u)>threshold:
            iii=iii*1.4
        else:
            b=iii
            bit=False
    
    ##########integrate##########
    E=np.linspace(a,b,N) #grit
    dE=(b-a)/N #step
    value= dE*np.sum(f(E,T,u)) #sum: step * grit values
    
    
    #print('integrate: \n b=',b,'\n integral=',value) #///testing
    
    return value

def nested_interval_u(f,T):
    #calculate u from f(E,T,u) by nested interval approximation
    
    ##########setup##########
    result=1 #expected integration result
    #interval:
    n=-5 #start
    m=5 #end
    uncertainty=1e-9
    
    ##########nested interval approx.##########
    i=0
    while i<200:
        w=(m+n)/2
        #print('w=',w) #///testing
        resultw=integrate(f,T,w)
        
        if resultw < 1:
            n=w
        elif resultw > 1:
            m=w
        else:
            n=w*99/100000
            m=w*101/100000
            
        i=i+1
        
        if np.abs(resultw-1) < uncertainty :
            i=99999999
    #print('result:\n T=',T,'\n µ=',w,'\n integral=',resultw,'\n done \n') #///testing
        
    return w

###defined functions:
#integrate(f,T,u)
#nested_interval_u(f,T)

T0=1e-6
#first solution
mu= nested_interval_u(fu,T0)
#print('\n\n\n','µ(T=',T0,')=',mu,'\n','integral=',integrate(fu,T0,mu),'\n\n\n')
T_F=mu


#obtain values from T0 to 2T_F with M steps
M=200
T=np.linspace(T0,2*T_F,M)
#T=np.logspace(np.log10(T0),np.log10(2*T_F),M)
mue=np.zeros(M)
for j in range(0,M):
    mue[j]=nested_interval_u(fu,T[j])


#internal energy
Ee =np.zeros(M)
for ii in range(0,M):
    Ee[ii]= integrate(fe,T[ii],mue[ii])

########################################plot


plt.figure(num=None, figsize=(10, 10), dpi=80, facecolor='w', edgecolor='k')

plt.subplot(211)
plt.scatter(T,mue, s=6, color='blue')

plt.title('chemisches Potential in Abhängigkeit von der Temperatur', fontsize=16)
plt.xlabel('T', fontsize=13)
plt.ylabel('µ(T)', fontsize=13)
plt.grid(True)

plt.subplot(212)
plt.scatter(T, Ee, s=6, color='blue')
           
plt.title('innere Energie in Abhängigkeit von der Temperatur', fontsize=16)
plt.xlabel('T', fontsize=13)
plt.ylabel('E(T)', fontsize=13)
plt.grid(True)

plt.tight_layout()
plt.show()
```
